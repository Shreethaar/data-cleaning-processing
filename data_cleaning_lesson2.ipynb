```python
# remove rows which has empty cells
new_df=df.dropna()
new_df
```

```python
print(new_df.to_string())
```

```python
# dropna() methods will returns a new DataFrame and will not manipulate the original
# if to change to original, simply use 'inplace=True' argument
df.dropna(inplace = True)
df
# the output will be the same as before but this case we did in the original dataset
```

```python
# there are other technique as well to replace null values, remove null entries isn't the best option
# we can fill it specifically using 'df.fillna({value],inplace=True})
df.fillna(130,inplace=True)
```

```python
# replaced only for specified columns
df['Calories'].fillna(130,inplace=True)
```

```python
# replaced using mean
x=df['Calories'].mean()
df['Calories'].fillna(x,inplace=True)
```

```python
# replaced using median
x=df['Calories'].median()
df['Calories'].fillna(x,inplace=True)
```

```python
# replaced using mode
x=df['Calories'].mode()[0]
df['Calories'].fillna(x,inplace=True)
```
