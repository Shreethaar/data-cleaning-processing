Methods for handling missing values:
1. Complete Case Analysis
- listwise deletion, remove any observation from dataset that contain missing values in any of the columns

2. Pairwise Deletion
- retain observations with missing values for specific analyses but only use the available data for each calculation
- allows maximize use of available data without discarding entire observations
- can lead to biased estimates if missingness is related to the variables being analyzed

3. Threshold-based Removal
- set a threshold for the proportion of missing values allowed in each observation or column. Observations which exceed threshold are removed
- allows the control of data loss by adjusting the threshold
- choosing right threshold can be subjective and require domain knowledge

4. Column-wise Removal
- If certain columns in your dataset have a high proportion of missing values and are not critical for your analysis, you may choose to remove those columns entirely.

5. Row-wise Removal
- missing values across multiple columns and cannot be imputed accurately, you may choose to remove those observations entirely.
- may reduce the sample size and statistical power of your analysis

6. Imputation before removal
-  imputing missing values for certain columns to retain as much information as possible.
-  allows you to balance the trade-off between data completeness and bias reduction

```python
numeric_columns = df.select_dtypes(include=['number']).columns

# Apply mean imputation to numeric columns
for column in numeric_columns:
    # Calculate the mean of the column
    column_mean = df[column].mean()
    # Fill missing values in the column with the mean
    df[column].fillna(column_mean)

df
```
